### Data reading

```python
vehicles = pd.read_csv("../Data/vehiculos_original.csv")
```

```python
vehicles.shape
```

```python
vehicles.head()
```

I'm going to rename the columns to give them a descriptive name

```python
vehicles = vehicles.rename(columns={ # Irename the columns to be legible
    "cylinders":"cilindros",
    "trany":"transmision",
    "make":"fabricante",
    "model":"modelo",
    "displ":"desplazamiento", #motor displacement volume
    "drive":"traccion",
    "VClass":"clase",
    "fuelType":"combustible",
    "comb08":"consumo", #combined MPG for fuelType1
    "co2TailpipeGpm":"co2", # tailpipe CO2 in grams/mile
})
```

```python
vehicles.head()
```

```python
vehicles.dtypes #lee los tipos de datos que tenemos, siempre no esta bien, hay que vigilar
```
