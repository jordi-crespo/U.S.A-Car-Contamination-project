###  Distribution of numerical variables

I'm going to use histograms to see the distribution of a variable

```python
vehicles['co2'].plot.hist();
```

Alternatively,it can be used the Kernel Density Estimate, 
which produces an estimated function of the distribution of the variable

```python
vehicles['co2'].plot.kde();
```

```python
def distribucion_variable_numerica(df, col):
    df[col].plot.kde()
    plt.xlabel('Distribucion de la variable {}'.format(col))
    plt.show()
```


Now I can use `pywidgets` to be able to analyze the distributions more comfortably. It makes sense to make these graphs only for numeric variables. The `pandas.select_dtypes` method returns those columns of specific types

```python

numerical_columns = vehicles.select_dtypes(['int', 'float']).columns
```

```python
from ipywidgets import interact, fixed
```

```python
interact(distribucion_variable_numerica, col=numerical_columns, df=fixed(vehicles));
```


it can be seen that the variable `cylinders` in spite of being numerical does not have a balanced distribution of values

```python
vehicles['cilindros'].value_counts(normalize=True)
```


It seems that a large part of the variables follow a normal distribution. That the variables follow a normal distribution is important since many algorithms assume that the distribution of the variables is normal.

We can check this with a [probability graph] (http://www.itl.nist.gov/div898/handbook/eda/section3/probplot.htm).

The function of scipy.stats `probplot` compares the distribution of a variable with a theoretical distribution (the default normal), the more the graph looks like a 45-degree line the more normal it will be.

```python
from scipy import stats
```

```python
def Numerical_variable_normality(col):
    stats.probplot(vehicles[col], plot=plt)
    plt.xlabel('Diagrama de Probabilidad(normal) de la variable {}'.format(col))
    plt.show()
```

```python
interact(Numerical_variable_normality, col=numerical_columns);
```


It seems that they follow a normal distribution, nevertheless, it is advisable to make sure doing a normality test. In a normality test, we want to reject the null hypothesis that the variable to be analyzed has been obtained from a population that follows a normal distribution. For a confidence level of 95%, we reject the null hypothesis if the p-value is less than 0.05. That is, if a P value (p-value) less than 0.05 is obtained, it means that the probabilities that the null hypothesis is true is so low (less than 5%) that we reject it.


scipy has the function [normaltest] (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html) that returns the p-value

```python
for num_col in numerical_columns:
    _, pval = stats.normaltest(vehicles[num_col])
    if(pval < 0.05):
        print("Column {} does not follow a normal distribution".format(num_col))
```

None of the numeric variables follow a correct normal distribution.


### Distribution of categorical variables

One way to see how the categorical variables are distributed is through the `pandas.value_counts` function. This function returns the number of existing records for each value of a column

```python
def categorical_variable_distribution(col):
    vehicles[col].value_counts(ascending=True,normalize=True).tail(20).plot.barh()
    plt.show()
```

```python
categorical_columns = vehicles.select_dtypes(['object', 'category']).columns
```

```python
interact(categorical_variable_distribution, col=categorical_columns);
```
